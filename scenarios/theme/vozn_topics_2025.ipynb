```python
import pandas as pd
import random
import re
import string
import numpy as np
from sklearn import decomposition
import matplotlib.pyplot as plt
```

```python
df_topics = pd.read_excel("/voz_TM_new.xlsx", sheet_name="Sheet1")
```

```python
df_tm = pd.read_excel("/voz_TM.xlsx", sheet_name="Sheet1")
```

```python
df_tm_new = pd.merge(df_tm, df_topics, how = "inner", left_index = True, right_index = True)
df_tm_new
```

```python
df_tm_new.drop(["Unnamed: 0_y", "Unnamed: 0_x", "Название", "Текст", "themes"], axis = 1, inplace = True)
df_tm_new
```

```python
df_tm_new.insert(loc=4, column='topics', value=df_tm_new['Темы'])
df_tm_new.drop(['Темы'], axis = 1, inplace = True)
df_tm_new
```

```python
df_tm_new = df_tm_new.rename(columns={0 : 'Домашняя', 1 : 'Failed', 2 : 'Мотив белого', 3 : 'Мотив неба', 4 : 'Морская', 5 : 'Музыкальная', 6 : 'Любовная', 7 : 'Мотив души', 8 : 'Молитва', 9 : 'Мотив устремления', 10 : 'Мотив снега', 11 : 'Мотив своего', 12 : 'Фигура поэта', 13 : 'Фигура женщины', 14 : 'Время', 15 : 'Мотив прошедшего', 16 : 'Мотив чувства', 17 : 'Мотив черного', 18 : 'Территориальная', 19 : 'Мотив сна', 20 : 'Мотив покидания/сохранения', 21 : 'Мотив полета', 22 : 'Мотив прощения', 23 : 'Мотив поля', 24 : 'Мотив длани'})
df_tm_new.head()
```

```python
TOPIC_NAMES = ["Домашняя", "Failed", "Мотив белого", "Мотив неба", "Морская", "Музыкальная", "Любовная", "Мотив души", "Молитва", "Мотив устремления", "Мотив снега", "Мотив своего","Фигура поэта", "Фигура женщины", "Время", "Мотив прошедшего", "Мотив чувства", "Мотив черного", "Территориальная", "Мотив сна", "Мотив покидания/сохранения", "Мотив полета", "Мотив прощения", "Мотив поля", "Мотив длани"]

for ind, row in df_tm_new.iterrows():
    max_val = row['max']
    max_topic = None
    for topic in TOPIC_NAMES:
      if row[topic] == max_val:
            max_topic = topic
    df_tm_new.loc[ind, 'max'] = max_topic

df_tm_new.head()
```
